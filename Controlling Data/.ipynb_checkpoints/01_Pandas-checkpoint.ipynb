# Pandas

```python
import pandas as pd
import numpy as np
```

데이터 유형 - Series, DataFrame, 날짜

```python
s = pd.Series([1,3,5,np.nan,6,8])
print(s)

dates = pd.date_range('20200215',periods=6)
print(dates)

df = pd.DataFrame(np.random.randn(6,4), index=dates, columns=['A','B','C','D'])
print(df)
```

```python
print(df.head(3))
print(df.index)
print(df.columns)
print(df.values)
print(df.info())
print(df.describe())
print(df.sort_values(by='B', ascending=False))
```

loc,iloc 함수

```python
print(df['A'])
print(df[1:3])
print(df['20130102':'20130104'])
#print(df[1:3,'A']) 오류
print("-"*30)

print(df.loc[dates[0]])
print(df.loc[:,['A','B']])
print(df.loc)
print("-"*30)

print(df.iloc[3])
print(df.iloc[[1,2,4],[0,2]])
print(df.iloc[:,1:3])
```

```python
print(df[df.A>0])
print(df[df>0])
```

데이터프레임 복사

```python
#df2= df ;데이터 위치가 복사되는 것으로 df2 변경시 df도 변경
df2= df.copy()
df2['E']=['one','one','two','three','four','three']
print(df2)
```

isin 함수

```python
df2[df2['E'].isin(['two','four'])]
```

apply 함수

```python
df.apply(np.cumsum)
```

```python
df.apply(lambda x: x.max()-x.min())
```

# Pandas 고급 기능

두 DataFrame 병합하기 - concat

```python
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']},
                   index=[0,1,2,3])
df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
                   'B':['B4','B5','B6','B7'],
                   'C':['C4','C5','C6','C7'],
                   'D':['D4','D5','D6','D7']},
                   index=[4,5,6,7])
df3 = pd.DataFrame({'A':['A8','A9','A10','A11'],
                   'B':['B8','B9','B10','B11'],
                   'C':['C8','C9','C10','C11'],
                   'D':['D8','D9','D10','D11']},
                   index=[8,9,10,11])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
result = pd.concat([df1,df2,df3])
result
```

```python
result = pd.concat([df1,df2,df3], keys=['x','y','z'])
result
```

```python
result.index
```

```python
result.index.get_level_values(0)
```

```python
result.index.get_level_values(1)
```

```python
df4 = pd.DataFrame({'B':['B2','B3','B6','B7'],
                   'D':['D2','D3','D6','D7'],
                   'F':['F2','F3','F6','F7']},
                   index=[2,3,6,7])
result = pd.concat([df1,df4],axis=1)
```

```python
result
```

```python
result = pd.concat([df1,df4], axis=1, join='inner')
result
```

```python
result = pd.concat([df1,df4], axis=1, join_axes=[df4.index])
result
```

```python
result = pd.concat([df1,df4], ignore_index=True)
result
```

두 DataFrame 병합하기 - merge

```python
left = pd.DataFrame({'key':['K0','K4','K2','K3'],
                    'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3']})
right = pd.DataFrame({'key':['K0','K1','K2','K3'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']})
pd.merge(left,right,on='key')
```

```python
pd.merge(left,right,how='left',on='key')
```

```python
pd.merge(left,right,how='right',on='key')
```

```python
pd.merge(left,right,how='outer',on='key')
```

```python
pd.merge(left,right,how='inner',on='key')
```
